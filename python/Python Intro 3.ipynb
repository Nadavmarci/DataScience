# Regular Expressions in Python
 
 



```python
import re
```

```python
str = '''
Name: Mr. Peter Johns
eMail: peter@gmail.com
Phone: (54)-333-25-22

Name: Mrs. Mary Green
eMail: mary@yahoo.com
Phone: (50)-231-00-19

Name: Mr. George Brown
eMail: gbrn@bbb.com
Phone: (49)-882-55-51
'''
```

```python
#str='abc@example.com'
match=re.search(r'\w+@\w+.\w+',str) #return abc@example.com

match
```

```python
#num='555-555-5555'
match_num=re.search(r'(\W\d{2}\W.\d{3}.\d{2}.\d{2})',str) 
match_num
```

```python
### Print eMails
pattern = re.compile(r'\w+@\w+.\w+')
matches = pattern.finditer(str)

matches

for match in matches:
    print(match)
```

```python
### Print phone numbers
pattern = re.compile(r'(\W\d{2}\W.\d{3}.\d{2}.\d{2})')
matches = pattern.finditer(str)

for match in matches:
    print(match)
```

```python
### Print the names of the males (Mr.) in the list
### The ? sign indicates that the point is optional
pattern = re.compile(r'Mr\.?\s[A-Z]\w*\s[A-Z]\w*')
matches = pattern.finditer(str)

for match in matches:
    print(match)
```

```python
### Print the names of the males (Mr.) in the list
### Print phone numbers
pattern = re.compile(r'(Mr|Mrs)\.?\s[A-Z]\w*\s[A-Z]\w*')
matches = pattern.finditer(str)

for match in matches:
    print(match)
```

```python
urls = '''
http://www.mysite.com
https://google.com/search?q=regular+expression+in+python
https://mail.google.com/
http://rstudio.somesite.org:8787/
http://yahoo.com
'''
```

```python
pattern = re.compile(r'https?://(\w+.)?\w+.(com|org)')
matches = pattern.finditer(urls)

for match in matches:
    print(match)
```

We can interact through the lines of the file and make some operations.
Lets divide the dataset into three independent datasets: one for each type of flower.

We will:
- Opend the files for read/write/append
- Iterate though lines
- Check the content of the line for the name of the flower
- Append the new data to the respective file
- Close the files


```python
### input file
f = open("iris.csv",'r')
```

```python
### output file
setosa = open("setosa.csv","a")
versicolor = open("versicolor.csv","a")
virginica = open("virginica.csv","a")
```

```python
### read the first line that has the column names
names = f.readline()
print(names)
```

```python
### write the column names into each of the three files
setosa.write(names)
versicolor.write(names)
virginica.write(names)
```

```python
iris = f.readlines()
for line in iris:
    if "setosa" in line:
        print("Setosa is in %s" % line)
        setosa.write(line)
    elif "versicolor" in line:
        print("Versicolor is in %s" % line)
        versicolor.write(line)
    else:
        print("Virginica is in %s" % line)
        virginica.write(line)
    
```

```python
### close the connection of the files
setosa.close()
versicolor.close()
virginica.close()
```

# Working with Pandas Data Frames

```python
import pandas as pd
 
iris = pd.read_csv("iris.csv")
iris.head()
```

```python
## add new colums to the df
iris['sepal_ratio'] = iris.sepal_length / iris.sepal_width
iris.head()
```

```python
iris['petal_ratio'] = iris.petal_length / iris['petal_width']
iris.head()
```

```python
## now proceed with checking the values of the df
iris.describe()
```

```python
## create categorical variables for the sepal and petal width
iris['petalw_cat'] = pd.np.floor(iris.petal_width)
iris['sepalw_cat'] = pd.np.floor(iris.sepal_width)
iris.describe()
```

```python
## subsetting
iris[iris.petalw_cat==2]
```

```python
iris.describe()
```

```python
### subsetting for given values
iris[iris.species.isin(['virginica','versicolor'])]
```

```python
# subsetting between values

iris2 = iris.copy()
iris2 = iris2[iris2['petal_length'].between(5,6)]
iris2.describe()
```
